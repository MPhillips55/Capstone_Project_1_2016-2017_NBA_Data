```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
pldf = pd.DataFrame.from_csv('combined-stats.csv', index_col=False)
```

```python
pldf.info()
```

```python
pldf.head()
```

```python
import time, datetime

def convertTime(t):
    x = time.strptime(t, '%H:%M:%S')
    return int(int(datetime.timedelta(hours=x.tm_hour,minutes=x.tm_min,seconds=x.tm_sec).total_seconds()))
```

```python
pldf['remaining_time_sec'] = pldf['remaining_time'].apply(convertTime)
```

```python
pldf.head()
```

```python
pldf['elapsed_sec'] = pldf['elapsed'].apply(convertTime)
```

```python
try:
    pldf['play_length_sec'] = pldf['play_length'].apply(convertTime)
except:
    pass
```

```python
pldf.play_length.replace(['00:-12:00', '00:-5:00'], ['00:00:00', '00:00:00'], inplace=True)
```

```python
pldf['play_length_sec'] = pldf['play_length'].apply(convertTime)
```

```python
pldf = pldf.rename(columns={'team_in_position':'team_in_possession'})
```

```python
pldf[['num', 'outof']] = pldf[['num', 'outof']].apply(pd.to_numeric) 
```

```python
pldf['num'] = pldf['num'].fillna(0).astype(int)
```

```python
pldf['outof'] = pldf['outof'].fillna(np.nan).astype(int)
```

```python
pldf['points'] = pldf['points'].fillna(0).astype(int)
```

```python
pldf.shot_distance.replace(['np.nan'], [np.nan], inplace=True)
```

```python
pldf[['shot_distance']] = pldf[['shot_distance']].apply(pd.to_numeric, errors='ignore') 
```

```python
pldf.original_x.replace(['unknown'], [np.nan], inplace=True)
pldf.original_y.replace(['unknown'], [np.nan], inplace=True)
pldf.converted_x.replace(['unknown'], [np.nan], inplace=True)
pldf.converted_y.replace(['unknown'], [np.nan], inplace=True)
```

```python
pldf[['original_x', 'original_y', 'converted_x', 'converted_y']] = pldf[['original_x', 'original_y', 'converted_x', 'converted_y']].apply(pd.to_numeric) 
```

```python
pldf.drop('data_set', axis=1,inplace=True)
```

```python
pldf.drop('play_length', axis=1,inplace=True)
pldf.drop('elapsed', axis=1,inplace=True)
pldf.drop('remaining_time', axis=1,inplace=True)
```

```python
pldf['date'] = pd.to_datetime(pldf['date'])
```

```python
pldf.head()
```

```python
pldf['game_id'] = pldf['game_id'].map(lambda x: x.lstrip('="').rstrip('"'))
```

```python
pldf['game_id'] = pldf['game_id'].fillna(0).astype(int)
```

```python
pldf.to_csv('play_by_play_cleaned.csv')
```
